```python
def get_upcs_first_page():
    # URL de la primera página
    url = "https://books.toscrape.com/catalogue/page-1.html"
    upcs = []

    # Solicitar la página principal con la lista de libros
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')

    # Encontrar todos los enlaces de libros en la página
    books = soup.select(".product_pod h3 a")

    # Bucle para cada libro
    for book in books:
        # Construir la URL para la página de detalles del libro
        book_url = "https://books.toscrape.com/catalogue/" + book["href"]
        
        # Solicitar la página de detalles del libro
        book_detail_page = requests.get(book_url)
        detail_soup = BeautifulSoup(book_detail_page.content, "html.parser")

        # Extraer el UPC
        upc = detail_soup.select_one("table.table.table-striped tr:nth-of-type(1) td").text
        
        # Almacenar el UPC y el título del libro
        upcs.append({
            "Title": book["title"],
            "UPC": upc
        })

    # Convertir a un DataFrame para un análisis más fácil
    return pd.DataFrame(upcs)

# Llamar a la función para obtener los UPCs de la primera página
upcs_df_first_page = get_upcs_first_page()

# Imprimir el DataFrame que contiene los UPCs y los Títulos
print(upcs_df_first_page)
```
